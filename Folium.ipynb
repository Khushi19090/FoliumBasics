```python
import folium
```

```python
m = folium.Map(location = [23.185884,79.974380])
m              
```

```python
folium.Marker([23.13158810059409, 79.797596388698],popup = "Bedaghat Waterfall").add_to(m)
m
```

```python
m = folium.Map(location = [23.185884,79.974380], zoom_start = 8)
folium.Marker([23.13158810059409, 79.797596388698],
               popup = 'Bhedaghar Waterfall',
               icon = folium.Icon(color = 'red', icon = 'glyphicon glyphicon-star')).add_to(m)
m
```

```python
m.add_child(folium.ClickForMarker(popup = 'New Places'))
```

```python
m = folium.Map(location = [23.185884,79.974380], zoom_start = 8)
folium.Circle(
    radius = 400,
    color= 'blue',
    location = [22.941382889997996, 79.92296950304208],
    fill = True
).add_to(m)

folium.CircleMarker(
    radius = 4,
    color= 'red',
    location = [22.94552656019422, 79.91789397956292],
    fill = True
).add_to(m)

m
```

```python
m = folium.Map(location = [23.185884,79.974380], zoom_start = 8) 
folium.TileLayer('Stamen Terrain').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('Stamen Water Color').add_to(m)
folium.TileLayer('cartodbpositron').add_to(m)
folium.TileLayer('cartodbdark_matter').add_to(m)
folium.LayerControl().add_to(m)
m
```
